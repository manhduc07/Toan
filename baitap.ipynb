```python
import numpy as np
import pandas as pd
```

```python
V = np.zeros((6, 12))
```

```python
V[0, 11] = 1
V[1, 11] = -1
gamma = 0.9
p = 0.8
p_down = 0.2
```

```python
lock = [(0, 4), (0, 8),
        (1, 4), (1, 6), (1, 8),
        (2, 1), (2, 2), (2, 4), (2, 6), (2, 8),
        (3, 1), (3, 6), (3, 8),
        (4, 1), (4, 2), (4, 3), (4, 4), (4, 5), (4, 6), (4, 8), (4, 9),
        (5, 4)]
```

```python
for i in lock:
  V[i] = -100
```

```python
for _ in V:
  print(_)
```

```python
def canMove(x, y, V, block_moves):
  up       = (x-1, y)
  down     = (x+1, y)
  left     = (x, y-1)
  right    = (x, y+1)
  moves    = [up, down, left, right]
  canMoves = []
  for move in moves:
      move_x, move_y = move
      if 0 <= move_x < V.shape[0] and 0 <= move_y < V.shape[1]:
            if move not in block_moves:
              canMoves.append(move)
  return canMoves
```

```python
def allMoves(V, block_moves):
  results = []
  for i in range(V.shape[0]):
    for j in range(V.shape[1]):
      obj = {(i, j): canMove(i, j, V, block_moves)}
      results.append(obj)
  return results;
```

```python
def addValue(value):
    allVal = []
    for val in value:
        allVal.append(val)
    return allVal
```

```python
def updateValue(value, gamma, V, p_down):
    listValue = addValue(value)
    maxValue = 0
    if len(value) > 1:
        sumVal = len(value) - 1
    else:
        sumVal = len(value)
    result = []
    for i in range(len(listValue)):
        sum = 0
        x_i, y_i = listValue[i]
        for j in range(len(listValue)):
            x_j, y_j = listValue[j]
            if j == i:
                sum += 0.8 * gamma * V[x_i][y_i]
            else:
                sum += (p_down / sumVal) * gamma * V[x_j][y_j]
        result.append(sum)
    maxValue = max(result)
    return maxValue
```

```python
moves = allMoves(V, lock)
```

```python
import pandas as pd

for __ in range(20):
  for i in moves:
    for _ in i:
      if(_ == (0,11) or _ == (1,11) or _ in lock):
         continue
      value = i[_]
      V[_] = updateValue(value, gamma, V, p_down)
  print(f'lan {__}')
  display(pd.DataFrame(V))
```
